## DepMap:

The goal is to parse gene_effect_corrected.csv file from DepMap (https://ndownloader.figshare.com/files/14221385)
and extract genes that are essential for cell viability. For this purpose, we find genes with effect score below a threshold which is -0.5 (https://depmap.org/portal/faq/#dep_thresholds).

```python
import csv
import wget
import pandas as pd
```

## Load the data

```python
url = 'https://ndownloader.figshare.com/files/14221385'
filename = '/Users/ravanv/PycharmProjects/IDGKG-gen/data/gene_effect_corrected.csv'
filename = wget.download(url)
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
columns = df.columns
df.shape
```

## Extract genes that have effect score below threshold (-0.5) on each cell line.

```python
gene_effect_cell_line_dict = {}
threshold  = -0.5 #TODO: need to check it.
row = 0
row_number = 21 #for test, otherwise row_number = df.shape[0]
col_number = 100 #for test, otherwise col_number = df.shape[1]
for i in range(1,row_number): 
    #print("i=",i)
    genes_effect = []
    for j in range(1,col_number):
        if df.loc[i][j] <= threshold: #TODO: need to check this condition
            #print(i,j,df.loc[i][j])
            genes_effect.append(columns[j])
    row += 1
    if row % 5 == 0:
            print("finished processing {} rows".format(row))       
    gene_effect_cell_line_dict[df.iloc[i][0]] =  genes_effect       
```

```python
print(gene_effect_cell_line_dict)
```

```python
for key in gene_effect_cell_line_dict.keys():
    if (gene_effect_cell_line_dict.get(key) != []):
        print(key,gene_effect_cell_line_dict.get(key) )
```
